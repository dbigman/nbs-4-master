```python
# Supervised Learning Class
# we will be using the pre loaded breast cancer dataset
from sklearn.datasets import load_breast_cancer
import numpy as np
cancer = load_breast_cancer()
```

```python
# description of the dataset
print(cancer['DESCR'])
```

```python
# 212 people with cancer
# 357 people without cancer
```

```python
#loading dataset to pandas
import pandas as pd

features = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
labels = pd.Series(cancer['target'], name = 'labels')
display(features.head(20))
display(labels.head(20))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, random_state=0)
```

# KNN (& overfitting)

```python
# let's do a KNN first
from sklearn.neighbors import KNeighborsClassifier

# initialize the model -> set hyperparameters
model = KNeighborsClassifier(n_neighbors = 9)
```

```python
# training your model on the training datasets -> where the algorithm will learn
model = model.fit(X_train, y_train)
```

```python
# once we have an algorithm that has learned, we can use it to predict on new instances:
model.predict(X_test)
```

```python
# and compare with the "real" y_test
np.array(y_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(model.predict(X_test),np.array(y_test))
```

```python
# the natural thing to do is to perform an evaluation of how well our model behaves with test data

# once the model is trained you can call the score method, to compare results of test predictions with actual values -> returns the accuracy
print("test data accuracy was ",model.score(X_test,y_test))

# you should always also see the accuracy of the training
print("train data accuracy was ", model.score(X_train, y_train))
```

## Overfitting basics

```python
#### WE WILL LEARN MORE ABOUT OVERFITTING IN A FUTURE LESSON BUT FOR NOW LET'S TAKE A LOOK ####

# These are actually quite good results, we don't seem to be overfitting
# let's see the test-training tradeoff as we force KNN to overfit
import matplotlib.pyplot as plt

# hyerparameter tuning
training_accuracy = []
test_accuracy = []
# try n_neighbors from 1 to 30
neighbors_settings = range(1, 30)
for n_neighbors in neighbors_settings:
  # build the model
  clf = KNeighborsClassifier(n_neighbors=n_neighbors)
  #training the model
  clf.fit(X_train, y_train)
  # record training set accuracy
  training_accuracy.append(clf.score(X_train, y_train))
  # record generalization accuracy
  test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()

plt.show()
# these types of charts are very important!! how does the test/training accuracy vary with a change in a hyperparameter
```

```python
#10 to 15 neighbours seems to be the optimal point
```

# Decision Trees (& Hyperparamenters & Feature importance)

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth = 10)
model.fit(X_train, y_train)
```

```python
#Trees are, unfortunately, overfitty
# relatively low performance on test
print("test data accuracy was ",model.score(X_test,y_test))
# 100% on training data
print("train data accuracy was ",model.score(X_train,y_train))
```

```python
# why dont I repeat the process but choose the max_depth of the tree
max_depth = range(1,30)
test = []
train = []

for depth in max_depth:
  model = DecisionTreeClassifier(max_depth= depth)
  model.fit(X_train, y_train)
  test.append(model.score(X_test,y_test))
  train.append(model.score(X_train,y_train))
```

```python
# we quickly see overfitting properties
import matplotlib.pyplot as plt
plt.plot( train, label="training accuracy")
plt.plot( test, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_depth")
plt.legend()
```

```python
# visualizing your decision trees
model = DecisionTreeClassifier(max_depth=3)
model.fit(X_train, y_train)

from sklearn.tree import export_graphviz
export_graphviz(model, out_file="tree.dot", class_names=["malignant", "benign"],
feature_names=cancer.feature_names, impurity=False, filled=True)

import graphviz
with open("tree.dot") as f:
  dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
# this code helps you plot the feature importance!
import numpy as np
def plot_feature_importances_cancer(model):
  n_features = cancer.data.shape[1]
  plt.barh(range(n_features), model.feature_importances_, align='center')
  plt.yticks(np.arange(n_features), cancer.feature_names)
  plt.xlabel("Feature importance")
  plt.ylabel("Feature")
plot_feature_importances_cancer(model)
```
